# Iteration in Python

Iteration means **repeating a set of instructions multiple times**.  
In Python, iteration is usually done using **loops**.

There are two main types of loops in Python:
1. `for` loop â†’ Used when we know the number of times we want to repeat.
2. `while` loop â†’ Used when we repeat until a condition becomes false.

Weâ€™ll explore each with examples.


### Example 1: Using a for loop with range()

````python
print("Counting from 1 to 5:")
for i in range(1, 6):
    print(i)
````

### Explanation
- `range(1, 6)` â†’ generates numbers `1,2,3,4,5`
- The loop assigns each number to `i`
- `print(i)` displays the current value of `i`


### Example 2: Iterating over a list

````python
fruits = ["apple", "banana", "cherry"]

print("List of fruits:")
for fruit in fruits:
    print(fruit)
````

### Explanation
- Here, instead of numbers, we are iterating over **elements of a list**.
- The loop variable `fruit` takes each value: "apple" â†’ "banana" â†’ "cherry".


### Example 3: Using while loop

````python
count = 1
while count <= 5:
    print("Count is:", count)
    count += 1   # increment the value
````

### Explanation
- The loop starts with `count = 1`.
- As long as `count <= 5` is true, the loop body executes.
- `count += 1` increases the value by 1 each time.
- Finally, when `count` becomes 6, the condition fails, and the loop stops.


### Example 4: Break and Continue

````python
print("Demonstrating break and continue:")

for i in range(1, 10):
    if i == 5:
        print("Breaking the loop at i =", i)
        break      # exits the loop completely
    if i % 2 == 0:
        print("Skipping even number:", i)
        continue   # skips to next iteration
    print("Processing number:", i)
````

### Explanation
- **break**: exits the loop completely.
- **continue**: skips the rest of the loop body and goes to the next iteration.
- Output flow:
  - Numbers 1 â†’ printed
  - 2 â†’ skipped
  - 3 â†’ printed
  - 4 â†’ skipped
  - 5 â†’ breaks the loop


### Example 5: Nested Loops

````python
print("Multiplication Table (1 to 3):")
for i in range(1, 4):
    for j in range(1, 4):
        print(f"{i} x {j} = {i*j}")
    print("------")
````

### Explanation
- A **nested loop** means one loop inside another.
- For each `i` value (1 â†’ 2 â†’ 3), the inner loop runs completely with `j` (1 â†’ 2 â†’ 3).
- This creates a multiplication table.


# Summary
- Iteration means repeating tasks.
- Two main loops in Python: `for` and `while`.
- `range()` is often used with `for` loops.
- Special keywords:
  - **break** â†’ exit loop
  - **continue** â†’ skip iteration
- Nested loops help solve complex problems (like tables, patterns).

ðŸ‘‰ Iteration is the foundation of repetitive tasks in programming.


# Classes and Objects

# Classes in Python

## What is a Class?
- A **class** is like a blueprint for creating objects.
- It defines **attributes (variables/data)** and **methods (functions/behaviors)**.
- An **object** is an instance of a class (something built using the class blueprint).

In simple terms:
- **Class â†’ Blueprint**
- **Object â†’ Real thing created from blueprint**


# Class Syntax in Python

The general syntax of a class in Python is:

```python
class ClassName:
    # class attribute (shared by all objects)
    class_variable = value
    
    # constructor (initializer)
    def __init__(self, parameters):
        self.instance_variable = parameters   # instance attributes

    # method (behavior)
    def method_name(self, parameters):
        # code for the method
        return something


# Summary of Classes in Python

- **Class** â†’ Blueprint for creating objects.
- **Object** â†’ Instance of a class.
- **Constructor (`__init__`)** â†’ Initializes object attributes.
- **Attributes** â†’ Variables inside class (instance or class variables).
- **Methods** â†’ Functions inside class (define behaviors).
- **Inheritance** â†’ One class can use properties/methods of another.
- **Encapsulation** â†’ Hiding private data from outside access.

ðŸ‘‰ Classes are the foundation of **Object-Oriented Programming (OOP)** in Python.


# Objects in Python

## What is an Object?
- An **object** is an instance of a class.
- Think of a **class** as a blueprint, and an **object** as a real-world thing created from that blueprint.
- Objects have:
  - **Attributes (data/variables)**
  - **Methods (functions/behaviors)**

ðŸ‘‰ In Python, **everything is an object** (numbers, strings, lists, user-defined classes, etc.).


# Syntax for Creating Objects in Python

## Class Definition
```python
class ClassName:
    def __init__(self, parameters):
        self.attribute1 = parameters
        self.attribute2 = parameters
    
    def method_name(self):

        return something


# Summary: Objects in Python

- **Class** = Blueprint
- **Object** = Instance created from class
- Objects have:
  - **Attributes** â†’ Variables inside an object
  - **Methods** â†’ Functions that belong to an object
- Objects are unique in memory (use `id()` to check).
- Objects allow us to model real-world things in code.

ðŸ‘‰ Without objects, classes are just definitions.  
ðŸ‘‰ Objects bring classes to life!


# Example of Class And Objects


### Example 1: Defining a simple class and objects


````python
class Car:
    # attribute (class variable)
    wheels = 4  

    # constructor method (called when object is created)
    def __init__(self, brand, color):
        self.brand = brand     # instance attribute
        self.color = color     # instance attribute

    # method (behavior of the object)
    def drive(self):
        print(f"The {self.color} {self.brand} is driving.")

# creating objects (instances)
car1 = Car("Tesla", "Red")
car2 = Car("BMW", "Black")

# accessing attributes
print("Car1 brand:", car1.brand)
print("Car2 color:", car2.color)

# calling methods
car1.drive()
car2.drive()
````

### Explanation
- `class Car:` â†’ defines a class named **Car**.
- `__init__` â†’ special **constructor method** that runs when an object is created.
- `self` â†’ refers to the object itself (needed to access attributes/methods).
- `car1 = Car("Tesla", "Red")` â†’ creates an object of class Car.
- `car1.brand` â†’ accesses the attribute of that specific object.
- `car1.drive()` â†’ calls the method of that object.


### Example 2: Adding methods to modify data

````python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def display(self):
        print(f"Student: {self.name}, Marks: {self.marks}")

    def update_marks(self, new_marks):
        self.marks = new_marks
        print(f"{self.name}'s marks updated to {self.marks}")

# create student object
s1 = Student("Alice", 85)
s1.display()

# update marks
s1.update_marks(92)
s1.display()
````

### Explanation
- The class `Student` has:
  - attributes: `name`, `marks`
  - methods: `display()`, `update_marks()`
- `s1.update_marks(92)` â†’ changes the marks of `s1`.


### Example 3: Class Variables vs Instance Variables

````python
class Dog:
    species = "Mammal"   # class variable (shared by all objects)

    def __init__(self, name, age):
        self.name = name   # instance variable (unique for each object)
        self.age = age

# create objects
dog1 = Dog("Tommy", 3)
dog2 = Dog("Bruno", 5)

print(dog1.name, "is a", dog1.species, "and is", dog1.age, "years old.")
print(dog2.name, "is a", dog2.species, "and is", dog2.age, "years old.")
````

### Explanation
- `species` â†’ a **class variable** (common to all dogs).
- `name` and `age` â†’ **instance variables** (different for each object).
- Both `dog1` and `dog2` share `species = "Mammal"`.


### Example 4: Inheritance (One class can use another class)

````python

class Animal:
    def speak(self):
        print("Animals make sounds")

class Cat(Animal):  # Cat inherits from Animal
    def speak(self):
        print("Meow Meow!")

class Dog(Animal):  # Dog inherits from Animal
    def speak(self):
        print("Woof Woof!")

# create objects
a = Animal()
c = Cat()
d = Dog()

a.speak()
c.speak()
d.speak()
````

### Explanation
- `class Cat(Animal):` â†’ means **Cat inherits from Animal**.
- If we donâ€™t redefine `speak()`, Cat would use Animalâ€™s method.
- But since Cat defines its own `speak()`, it overrides the parentâ€™s method.


### Example 5: Encapsulation (Restricting Access)

````python
class BankAccount:
    def __init__(self, account_holder, balance):
        self.account_holder = account_holder
        self.__balance = balance   # private variable (hidden outside the class)

    def deposit(self, amount):
        self.__balance += amount
        print(f"Deposited {amount}. New Balance: {self.__balance}")

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew {amount}. Remaining Balance: {self.__balance}")
        else:
            print("Insufficient funds!")

# create account
acc = BankAccount("John", 1000)

acc.deposit(500)
acc.withdraw(200)

# Trying to access private variable directly
# print(acc.__balance)  # âŒ Error: cannot access private variable
````

### Explanation
- `__balance` â†’ is a **private attribute** (cannot be accessed directly outside the class).
- Access is only possible using methods (`deposit` and `withdraw`).
- This is called **encapsulation** (data hiding).
