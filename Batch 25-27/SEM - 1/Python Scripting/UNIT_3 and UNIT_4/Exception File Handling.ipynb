EXCEPTION FILE HANDLING

```python
#Exception handling for file I/O involves using try, except, else, and finally blocks to manage potential errors
#FileNotFoundError: Raised when attempting to open a file that does not exist
#PermissionError: Raised when attempting to access a file without the permission
```

```python
#FileNotFoundError
try: 
    with open("Josh.txt", "r") as file:
         content = file.read()
    print ("file content:", content)
except FileNotFoundError:
    print("Error: The file 'Josh.txt' does not exist.")
```

```python
#PermissionError 
try: 
    with open("Josh.txt", "r") as file:
         content = file.read()
    print ("file content:", content)
except IOError:
    print("An unexpected error occurred.")
except PermissionError:
    print("Error: Unable to access 'Josh.txt'.")
```

```python
#Finally block is optional and executes regardless of wether an exception occurred
try: 
    with open("Josh.txt", "r") as file:
         content = file.read()
    print ("file content:", content)
except IOError:
    print("An unexpected error occurred.")
except PermissionError:
    print("Error: Unable to access 'Josh.txt'.")
finally:
    print("Attempted file operation.")
```
