# **Advantages of Python**

#### **It is easy to read, learn, and write :** Its simplicity and readability, makes it an excellent choice for both beginners and experienced programmers.
#### **Object-Oriented and Procedural programming language :** It is a multi-level programming languuage where we can write purely procedural code or uses classes/objects, or mix both.
#### **Extensive support libraries :** It has extensive support libraries like NumPy for numerical calculations and Pandas for data analytics, making it suitable for scientific and data-related applications.
#### **Cross-Platform Compatibility :** Works across multiple diffrent operating systems, allowing real-time code execution and testing.
#### **Integration with other languages :** It can work together with code written in other languages such as C or C++. Giving us more flexibility.

# **Disadvantages of Python**

#### **Performance :** Its an interpreted language, which means that it can be slower than compiled languages like C or Java. This can be an issue for performance-intensive tasks.
#### **Memory Overhead :** It can consume a lot of memory as compaired to some lower leveled languages, especially when working with large datasets or running complex algorithms.
#### **Steep learning curve :** While it is generally considered to be a relatively easy language to learn, it can still have a steep learning curve for beginners, especially if they have no prior experience with programming.
#### **Limited Support for Mobile Development :** It is not sutaible to use for mobile app development as it has relatively slower execution speeds and a higher memory consumption.
#### **Backward Compatability Issues :** Transitiong from one version to another can pose a significant challange for example when they stopped support for python 2 transitioning to pyhon 3 faced many challanges

# **Variable**

In python variables are elements that are used to store data such that it can be referenced and manipulated when executing a prgram. Unlike many other programming languages, Python variables do not require explicit declaration of type. The type of the variable is based on the value assigned.

```python
# Variable 'x' stores the integer value 10
x = 10

# Variable 'name' stores the string "Samantha"
name = "Mayank"  

# Printing the variable 
print(x)
print(name)
```

## **Rules**

#### There are certain rules we have to follow when creating variables such as:
- Names can only contain letters, digits and underscores
- Variable name cannot start with a digit
- Names are case-sensitive
- Avoid using Python keywords (e.g., if, else, for) as variable names

```python
# Valid Examples
age = 21
_colour = "lilac"
total_score = 90
```

```python
#n Invalid Exampkes
1name = "Error"  # Starts with a digit
class = 10       # 'class' is a reserved keyword
user-name = "Doe"  # Contains a hyphen
```

## **Type of Variable**

We can determine the type of a variable using the type() function. This built-in function returns the type of the variable.

```python
# Define variables with different data types
n = 42
f = 3.14
s = "Hello, World!"
li = [1, 2, 3]
d = {'key': 'value'}
bool = True

# Get and print the type of each variable
print(type(n))   
print(type(f)) 
print(type(s))   
print(type(li))     
print(type(d))     
print(type(bool))
```

### **Runtime Error**

These errors occur when the program is executing but crashes or behaves unexpectedly because of these errors. These are also called exceptions.

```python
# Zero Division Error

x = 10
y = 0

result = x / y # dividing variable x and y

print(result)
```

```python
# Debugging Zero Division Error using try-except

try: # the try block tries to run the code and if encounters an error runs the except block
    x = 10
    y = 0

    result = x / y

    print(result)

except ZeroDivisionError:
    print("Error: Division by zero not allowed")
```

```python
# Name Error

print(a)
```

```python
# Debugging Name Error

try:
    print(a)

except NameError:
    print("Error: Variable is not defined")
```
