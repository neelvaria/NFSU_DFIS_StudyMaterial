# STRINGS

```python
Python_Story=(
    "The Tale of the Messy Data. Once upon a time, in the digital kingdom of Data Processing,"
"lived a brilliant but overworked analyst named Alice. Her task was to manage the grand User Database,"
"but she faced a chaotic problem: the data was messy."
)
# Display original String
print("Your Original String:")
print("=====================")
print (Python_Story)
print("")

#The length of the string
print(f"Your first string size is : {len(Python_Story)} ")
print("")

#Replace the name Alice with Praveen
new_name="Praveen"
new_Python_Story= Python_Story.replace("Alice", new_name)

print("Your String with Alice Replaced by Praveen:")
print("===========================================")
print(new_Python_Story)
print("")


Other_Story="Once upon a time, in the land of the computer's memory, there was an intrepid adventurer named Variable."

# Display the second String
print("Your Second String:")
print("==================")
print (Other_Story)
print("")
#The length of the second string
print(f"Your Second string size is :{len(Other_Story)}")
print("")

#str.join()
Joint_String=", ".join([Python_Story, Other_Story])
print("Your First and Second String combined:")
print("=====================================")
print(Joint_String)
print("")
#The length of the combined string
print(f"Your Combined string size is : {len(Joint_String)}")
```

# SET
